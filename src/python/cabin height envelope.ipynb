```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(num=None, figsize=None, dpi=600, facecolor='w', edgecolor='k')
plt.plot([0, 6e3, 25e3, 27e3],
         [0, 6e3,  6e3, 27e3-19e3],'k')
#plt.axis('equal')
plt.axis([0,27e3,0,10e3])
plt.minorticks_on()
plt.grid()
plt.title('Envelope de pressurização da cabine')
plt.xlabel('Altitude de voo (ft)')
plt.ylabel('Altitude de cabine (ft)')
plt.savefig('../../images/cabin_envelope.png')
```
